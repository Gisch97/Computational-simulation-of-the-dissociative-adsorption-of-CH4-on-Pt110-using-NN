```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('classic')
sns.set_theme(style="whitegrid")
paletteE =['#00377e', '#006B8F', '#06889B', '#FEAE6C', '#FE6C31', '#F93414']
paletteT =['#011F51', '#06889B',  '#FE6C31']

sns.set_palette(palette=paletteE)
X = [ 0.6, 1,     1.4]
Ystick = [ 0.0123 ,0.069, 0.148]
Yad = [ 0.224,  0.05,  0.01 ]
```

```python
fig, ax = plt.subplots(1,2, figsize = (16,6))
sns.lineplot(data=MD, x='1/T', y='Prefl', hue='Ei', marker='o', ax=ax[0], palette=paletteE)
fig.suptitle('Probabilidad de reflexi√≥n molecular')
ax[0].legend(bbox_to_anchor=(-0.1,1), title='Ei[eV]')
ax[0].set_xlabel('1 / Temperatura [$K^{-1}$]')
ax[0].set_xlim(0, 22)
sns.lineplot(data=MD, x='Ei', y='Prefl', hue='T', marker='o', ax=ax[1], palette=paletteT)
ax[1].set_xlabel(f'Energ√≠a de incidencia [eV]')
ax[1].set_xlim(0.35, 1.45)
ax[1].legend(bbox_to_anchor=(1.2,1), title='T [K]')
```

```python
fig, ax = plt.subplots(1,2, figsize = (16,6))
sns.lineplot(data=MD.query('Ei>0.4'), x='1/T', y='Pstick', hue='Ei', marker='o', ax=ax[0], palette=paletteE)
fig.suptitle('Probabilidad de sticking', fontsize=16)
ax[0].legend(bbox_to_anchor=(-0.1,1), title='Ei[eV]')
ax[0].set_xlabel('1000 / $T_{sf}$ [$K^{-1}$]')
ax[0].set_xlim(0, 22)
sns.lineplot(data=MD.query('Ei>0.4'), x='Ei', y='Pstick', hue='T', marker='o', ax=ax[1], palette=paletteT)
ax[1].set_yscale('log')
ax[1].set_xlabel(f'Energ√≠a de incidencia [eV]')
ax[1].set_xlim(0.55, 1.45)
ax[1].set_ylim(0.005, 0.3)
ax[1].grid(which='minor')
ax[1].minorticks_on()
ax[1].legend(loc='upper left', title='T [K]')
fig.tight_layout()
plt.show()
```
